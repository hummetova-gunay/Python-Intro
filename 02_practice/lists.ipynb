```python
# Change the second item:

fruits = ["apple", "banana", "cherry"]
fruits[1] = "pear"
print(fruits)
```

```python
# Change the second value by replacing it with two new values:

fruits[1:2] = ["blackcurrant", "watermelon"]
print(fruits)
```

```python
# add items
fruits.append("orange")
print(fruits)
```

```python
# insert items
fruits.insert(1, "kiwi")
print(fruits)
```

```python
# remove items
fruits.remove("kiwi")
print(fruits)
```

```python
# remove items by index with the pop() method:
fruits.pop(1)
print(fruits)
```

```python
# remove items by index with the del statement:
del fruits[1]
print(fruits)
```

```python
# sort lists numerically
num_list = [5, 2, 9, 1, 5, 6]
num_list.sort()
print(num_list)
```

```python
# sort lists alphabetically
fruits = ["banana", "apple", "cherry"]
fruits.sort()
print(fruits)
```

```python
# sort descending
num_list.sort(reverse=True)
print(num_list)
```
