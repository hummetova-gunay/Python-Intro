```python
nums_list = [1,2,3,4,5,6]
even_list = []
for i in nums_list:
    if i%2 == 0:
        even_list.append (i)
```

```python
odd_list = []
for i in nums_list:
    if i%2 == 1:
        odd_list.append (i)
```

```python
duplicated_list = [1,1,2,4,4,4,5]
unique_list = []
for i in duplicated_list:
    if i not in unique_list:
        unique_list.append(i)
print(unique_list)
```

```python
for i in duplicated_list[:]:
    if duplicated_list.count(i) > 1:
        duplicated_list.remove(i)
print(duplicated_list)
```
