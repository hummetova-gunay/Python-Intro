```python
# Write a function that returns the square of a number.
def calc_square(a):
    return a**2
calc_square(3)
```

```python
# Write a function average(a, b, c) that returns the average of three numbers.
def avg(a,b,c):
    return (a+b+c)/3
avg(1,2,3)
```

```python
# Write a function is_even(n) that returns True if the number is even and False otherwise.
def is_even(num):
    return num%2==0
is_even(3)
```

```python
def check_num(num):
    if num%2 == 0:
        return True
    else:
        return False
check_num(3)
```

```python
# Write a function min_max(numbers) that takes a list and returns the smallest and largest number
def min_max(numbers):
    if not numbers:
        raise ValueError('Number must not be empty')
    min_num = min(numbers)
    max_num = max(numbers)
    return (min_num, max_num)
min_max([1,4,5,6,8])
```

```python
# Write a function sum_all(*numbers) that returns the sum of any number of inputs.
def sum_all(*numbers):
    return sum(numbers)
sum_all(1,2,3,4)
```

```python
# Write a function multiply_and_sum(multiplier, *numbers) that: Multiplies each number by multiplier and Returns the sum of the results.
def multiply_and_sum(multiplier, *numbers):
    # num_list = []
    # for num in numbers:
    #     num_list.append(multiplier*num)
    return sum(multiplier*num for num in numbers)

multiply_and_sum(2, 3,4,5)
```

```python
# Write a function: top_students(limit, *names), return the first limit students.
def top_students(limit, *names):
    return names[:limit]
```

```python
# Create a decorator print_start_end that prints:starting function and finished function before and after any function runs

def print_start_end(func):
    def wrapper():
        print('start')
        result = func()
        print('end')
        return result
    return wrapper
@print_start_end
def my_func():
    print('hello world')
my_func()
```

```python
# Write a decorator timer that measures how long a function takes to run.

import time
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'Execution time {end-start}')
        return result
    return wrapper

@timer
def my_func():
    time.sleep(5)
my_func()
```

```python
# 10. Write a function: multiply_all(*numbers) Return the product of all numbers.
def multiply_all(*numbers):
    product = 1
    for i in numbers:
        product*=i
    return product
multiply_all(1,2,3,4)
```

```python
import math
def multiply_all(*numbers):
    return math.prod(numbers)
multiply_all(1,2,3,4)
```

```python
# 11.Write a function: count_args(*args) Return how many arguments were passed.
def count_args(*arg):
    return len(arg)
count_args(1,2,4)
```
