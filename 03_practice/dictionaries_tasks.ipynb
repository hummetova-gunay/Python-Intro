```python
# 1.Create a dictionary storing 3 countries and their capitals.

countries_dict = {
    'Azerbaijan':'Baku',
    'UK':'London',
    'France':'Paris'
}
print(countries_dict)
```

```python
# 2.Write a function that prints all keys in a dictionary.
def get_keys(my_dict):
    for key in my_dict:
        print(key)

get_keys(countries_dict)
```

```python
# 3.Write a function that returns the sum of all values in a dictionary.
def sum_values(my_dict):
    sum =0
    for value in my_dict.values():
        sum+=value
    return sum
```

```python
# 4.Write a function that finds the key with the highest value.
def highest_val(my_dict):
    num_list = []
    for i in my_dict.values():
        num_list.append(i)
    return max(num_list)
```

```python
# 5.Count how many times each letter appears in a word using a dictionary.
word = 'Mahammad'
letters =[]
for i in word.lower():
    letters.append(i)
print(letters)



counts ={}
for letter in letters:
    counts[letter] = counts.get(letter, 0) +1
print(counts)
```
