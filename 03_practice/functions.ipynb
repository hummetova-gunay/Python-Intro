```python
# number of arguments - *args becomes tuple
def best_students(*students):
    print(type(students))
    print(f'The best students are {students[:2]}')
```

```python
best_students('Gunay', 'Nezrin', 'Aygun', 'Fatime', 'Melek')
```

```python
# combining regular and *args
def student_ages(age, num_students, *names):
    for name in names[0:num_students]:
        print(f'{name} is {age} years old')
```

```python
student_ages(21, 3, 'Gunay', 'Humay', 'Nezrin', 'Fidan')
```

```python
# decorators
def deco(func):
    def wrapper():
        print('Frist Hello')
        func()
        print('Second Hello')
    return wrapper
```

```python
@deco
def say_hello():
    print('Main Hello')
say_hello()
```

```python
def second_decorator(func):
    def wrapper(*args):
        print('Function is about to run')
        result  = func(*args)
        print('Finished running')
        return result
    return wrapper
```

```python
@second_decorator
def my_func(a,b):
    return a+b
my_func(2,4)
```
