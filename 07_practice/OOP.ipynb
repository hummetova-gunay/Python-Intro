## Class və object nədir?
- `Class` - şablon (məs. maşınlar)
- `Object` - bu şablon əsasında yaradılmış nümunə (məs. ford)
- `attiribute` - obyektin xüsusiyyətləri (məs. rəng, marka və s.)
- `method` - obyektin edə biləcəyi şeylər (məs. signal vermək, dayanmaq)
- `__init__` - constructor deyilir və obyekt yaradıldıqda avtomatik çağırılır
- `self` - obyektin özünə istinad edir

```python
class Car: # class
    def __init__(self, brand, color, year):
        self.brand = brand # attribute
        self.color = color # attribute
        self.year = year # attribute
        self.speed = 0
    def get(self): # method
        self.speed+=20
        print(f"{self.brand} is moving! current speed {self.speed}")
    def stop(self): # method
        self.speed = 0
        print(f'{self.brand} is stopping...')
    def info(self): # method
        print(f'The brand is {self.brand}, the color is {self.color}, the year is {self.year}')
```

## Inheritance

- `inheritance` - bir `class` başqa bir `class`ın xüsusiyyətlərini miras alır və üstünə yenilərini əlavə edir
- `Parent class` - Heyvan
- `Child class` - it, pişik və s.
- `super` - parent class-ın metodunu çağırmaq üçün istifadə olunur

```python


class Animal: # base class
    def __init__(self, name):
        self.name = name
    def speak(self):
        return f"{self.name} makes sound"
```

```python
class Dog(Animal): # child class
    def speak(self):
        return f'{self.name} barks'
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def info(self):
        print(f'Name is {self.name}, age is {self.age}')
    def make_sound(self):
        print(f'Animal makes sound ...')
```

```python
class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age) # parent class-dan __init__ metodunu cagir
        self.breed = breed # ozune xas olan attribut

    def make_sound(self):
        # return super().make_sound()
        print(f'{self.name} barks') # override

    def info(self):
        super().info() #override + expand
        print(f'Breed is {self.breed}')
```

```python
dog1 = Dog('Sofi', 3, 'Golden Retriever')
dog1.info()
```
