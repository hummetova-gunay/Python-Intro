# Two sum
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

```python
# two sum
def two_sum(num_list, target=0):
    indices = []
    for num in num_list: 
        for el in num_list:
            if num!=el and target == num+el:
                indices.append([num_list.index(num), num_list.index(el)])
    return indices
```

```python
two_sum([1,2,3], target=5)
```

```python
# brute force
def two_sum(num_list, target):
    for i in range(0, len(num_list)):
        for j in range(i+1, len(num_list)):
            if num_list[i]+num_list[j]==target:
                return [i,j]
```

```python
two_sum([1,2,3], target=5)
```

```python
# hash map
def hash_map(nums, target):
    seen ={}
    for i, num in enumerate(nums):
        complement = target-num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
```

```python
hash_map([1,2,3], 5)
```

# Add Two numbers
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

```python
def add_two_nums(l1, l2):
    l1_reversed = l1[::-1]
    l2_reversed = l2[::-1]
    sum_list = [a+b for a, b in zip(l1_reversed, l2_reversed)]
    return sum_list[::-1]
```

```python
add_two_nums([2,4,3], [5,6,4])
```

```python
# correct solution

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_two_nums(l1, l2):
    dummy = ListNode()
    current = dummy
    carry = 0

    while l1 or l2 or carry:

        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0

        total = v1 + v2 + carry
        carry = total // 10

        current.next = ListNode(total % 10)
        current = current.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy.next


def build_linked_list(values):
    dummy = ListNode()
    current = dummy

    for v in values:
        current.next = ListNode(v)
        current = current.next

    return dummy.next

def print_list(node):
    res = []
    while node:
        res.append(node.val)
        node = node.next
    return res
```

```python
l1 = build_linked_list([2,4,3])
l2 = build_linked_list([5,6,4])
result = add_two_nums(l1, l2)

print(print_list(result))
```
