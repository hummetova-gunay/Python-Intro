# 1. Check same case
Write a function that will check if two given characters are the same case. 
- If either of the characters is not a letter, return -1
- If both characters are the same case, return 1
- If both characters are letters, but not the same case, return 0. 

Examples 

- 'a' and 'g' returns 1

- 'A' and 'C' returns 1

- 'b' and 'G' returns 0


```python
def checkCase(s1:str, s2: str)->bool:
    """takes strings as an input and returns True if both lowercase or uppercase, False otherwise"""
    if not s1.isalpha() or not s2.isalpha():
        return -1
    if (s1.islower() and s2.islower()) or (s1.isupper() and s2.isupper()):
        return 1
    else:
        return 0
```

# 2. altERnaTIng cAsE <=> ALTerNAtiNG CaSe
Define a function such that each lowercase letter becomes uppercase and each uppercase letter becomes lowercase. For example:
- "hello world".toAlternatingCase() === "HELLO WORLD"
- "HELLO WORLD".toAlternatingCase() === "hello world"
- "hello WORLD".toAlternatingCase() === "HELLO world" 
- "12345".toAlternatingCase()       === "12345" 

```python
def alternateCase(s: str)->str:
    res = ''
    for letter in s:
        if not letter.isalpha():
            res += letter
            continue
        if letter.islower():
            letter = letter.upper()
        else:
            letter = letter.lower()
        res+= letter
    return res
```

```python
def alterCase(s: str)->str:
    return s.swapcase()
```

# 3. Alternating Loops
Write a function that combines arrays by alternatingly taking elements passed to it. Example
- combine(['a', 'b', 'c'], [1, 2, 3]) == ['a', 1, 'b', 2, 'c', 3]
- combine(['a', 'b', 'c'], [1, 2, 3, 4, 5]) == ['a', 1, 'b', 2, 'c', 3, 4, 5]
- combine(['a', 'b', 'c'], [1, 2, 3, 4, 5], [6, 7], [8]) == ['a', 1, 6, 8, 'b', 2, 7, 'c', 3, 4, 5]

```python
def combine(*lists):
    res =[]
    max_length = max(len(lst) for lst in lists)
    for i in range(max_length):
        for lst in lists:
            if i < len(lst):
                res.append(lst[i])
    return res
```

```python
combine(['a', 'b', 'c'], [1,2,3], ['l', 'm', 'n', 'o'])
```

# 4. What dominates your array
A zero-indexed array arr consisting of n integers is given. The dominator of array arr is the value that occurs in more than half of the elements of arr.
For example, consider array arr such that arr = [3,4,3,2,3,1,3,3]
The dominator of arr is 3 because it occurs in 5 out of 8 elements of arr and 5 is more than a half of 8.

```python
from collections import Counter
def findDominant(lst):
    counts = Counter(lst)
    value, freq = counts.most_common(1)[0]
    return value if freq > len(lst)/2 else -1
```

```python
findDominant([3,3,3,3,3,3,4,5,6,7,2])
```

# 5. Plus-minus-plus-minus Count
How often sign changes in list
lst = [1, -3, -4, 5]; ->     count -> 0,1,1,2
catchSignChange(lst) = 2;

```python
def catchSignChange(lst: list[int]) -> int:
    count = 0
    for i in range(len(lst) - 1):
        if (lst[i] > 0 and lst[i+1] < 0) or (lst[i] < 0 and lst[i+1] > 0):
            count += 1
            
    return count
```

```python
catchSignChange([1,-3, -4, 5])
```

```python
def catchSignChange(lst):
    return sum(
        (a > 0 and b < 0) or (a < 0 and b > 0)
        for a, b in zip(lst, lst[1:])
    )
```

## 1. Count uppercase letters

Write a function that returns how many uppercase letters are in a string.
- countUpper("HeLLo") == 3
- countUpper("abc") == 0

```python
def countUpper(s):
    count = 0
    for letter in s:
        count += letter.isupper()
    return count
```

```python
def countUpper(s: str) -> int:
    return sum(letter.isupper() for letter in s)
```

## 2. Replace negative numbers with 0
Write a function that takes a list of integers and returns a new list where every negative number is replaced by 0.
- fixNegatives([1, -2, 3, -5, 0]) == [1, 0, 3, 0, 0]

```python
def fixNegatives(lst: list):
    res = []
    for i in lst:
        if i>0:
            res.append(i)
        else:
            res.append(0)
    return res
```

```python
def fixNegatives(lst):
    return [0 if x<0 else x for x in lst]
```

## 4. Sum only even numbers

Write a function that returns the sum of all even numbers in a list.
- sumEven([1, 2, 3, 4, 5, 6]) == 12

```python
def sumEven(lst:list[int])->int:
    return sum(x if x % 2==0 else 0 for x in lst)
```

## 5. Double every second element

Return a new list where every element at an odd index is doubled.
- doubleSecond([1, 2, 3, 4, 5]) == [1, 4, 3, 8, 5]

```python
def doubleSecond(lst):
        return [lst[i] if i%2 == 0 else lst[i]*2 for i in range(len(lst))]
```

## 6. Find the first non-negative number

Return the first number in the list that is >= 0. If none exists, return -1.
- firstNonNegative([-5, -3, 0, 2]) == 0
- firstNonNegative([-4, -2]) == -1

```python
def firstNonNegative(lst):
    for x in lst:
        if x>=0:
            return x 
    return -1
```

- countValue([1, 2, 1, 3, 1], 1) == 3

```python
def countValue(lst, num):
    return Counter(lst)[num]
```
