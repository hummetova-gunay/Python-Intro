## Task 9 — True or False?

Predict the output of each line before running. Write your guess as a comment.

````python
# print(10 > 5)        # my guess: ?
# print(3 == 3.0)      # my guess: ?
# print("a".upper() == "A")   # my guess: ?
# print(10 != 10)      # my guess: ?
# print(bool(0))       # my guess: ?
print(bool("hello")) # my guess: ?
````

<details>
<summary>💡 Solution</summary>

```
10 > 5        → True
3 == 3.0      → True   (int and float are compared by value)
"a" == "A"   → False  (Python is case-sensitive)
10 != 10      → False  (10 is equal to 10)
bool(0)       → False  (0 is falsy)
bool("hello") → True   (non-empty string is truthy)
```
</details>
